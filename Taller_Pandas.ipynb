# Ejercicio 1 - Crear un DataFrame

En este ejercicio se creará un DataFrame utilizando datos almacenados en un diccionario.

```python
# Importar librerías

import pandas as pd
import numpy as np

# Crear un DataFrame con datos de ejemplo
datos = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Marta', 'Jorge'],
    'Edad': [23, 34, 29, 31, 28],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Bilbao'],
    'Puntuación': [85, 92, 78, 88, 95]
}

df = pd.DataFrame(datos)
print("DataFrame creado:")
df
```

# Ejercicio 2 - Operaciones entre columnas

Se realizarán operaciones matemáticas entre columnas para calcular el total de ventas.

```python

# Crear datos

ventas = {
    "Producto": ["Mouse", "Teclado", "Monitor"],
    "Precio": [50000, 80000, 700000],
    "Cantidad": [3, 2, 1]
}

# Crear DataFrame

df_ventas = pd.DataFrame(ventas)

# Crear nueva columna

df_ventas["Total"] = df_ventas["Precio"] * df_ventas["Cantidad"]

# Mostrar resultado

print(df_ventas)
```

# Ejercicio 3 - Lectura de archivos CSV

En este ejercicio se leerá un archivo CSV desde una fuente externa utilizando la función `read_csv()` de Pandas.

Se utilizará el dataset Titanic.

```python

# Leer dataset desde una URL

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df_titanic = pd.read_csv(url)

# Mostrar las primeras 10 filas del dataset

print("Primeras 10 filas del dataset Titanic:")

df_titanic.head(10)
```

# Ejercicio 4 - Datos sintéticos

Se crearán datos aleatorios utilizando NumPy para simular información.

```python

# Generar 100 registros sintéticos
np.random.seed(42)  # para reproducibilidad
sintetico = {
    'id': range(1, 101),
    'edad': np.random.randint(18, 70, size=100),
    'ingreso': np.random.normal(3000, 800, size=100).round(2),
    'compras_mes': np.random.poisson(5, size=100)
}

df_sint = pd.DataFrame(sintetico)
print("Primeras 10 filas de datos sintéticos:")
df_sint.head(10)
```
